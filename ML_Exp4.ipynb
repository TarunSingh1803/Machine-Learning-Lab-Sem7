```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df=pd.read_csv("Datasets\iris.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
X=df.iloc[:,:-1].values
```

```python
y=df.iloc[:,4].values
```

```python
# Split the dataset into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
# Standardize the features
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# Apply KNN classifier
KNN1=KNeighborsClassifier(n_neighbors=1, metric='euclidean')
KNN1.fit(X_train,y_train)
```

```python
# next step: Make predictions on the test set
y_pred1 = KNN1.predict(X_test)
```

```python
# Evaluate the model
accuracy1 = accuracy_score(y_test, y_pred1)
print("Accuracy:", accuracy1)
conf_matrix1 = confusion_matrix(y_test, y_pred1)
print("Confusion Matrix:\n", conf_matrix1)
print("Classification Report:\n", classification_report(y_test, y_pred1))
```

```python
# give predictions
predictions1 = y_pred1
print("Predictions:", predictions1)
```

```python
from sklearn.model_selection import cross_val_score

# Comparison of KNN Classifier with Different Distance Metrics
metrics = ['euclidean', 'manhattan', 'chebyshev', 'minkowski']
k_values = [1, 3, 5, 7, 9]

results = {metric: [] for metric in metrics}
X_scaled = scaler.transform(X)

for metric in metrics:
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, metric=metric)
        scores = cross_val_score(knn, X_scaled, y, cv=5)
        results[metric].append(scores.mean())

# Bar chart
bar_width = 0.2
index = np.arange(len(k_values))

plt.figure(figsize=(10,6))
for i, metric in enumerate(metrics):
    plt.bar(index + i*bar_width, results[metric], bar_width, label=metric)

plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Cross-Validated Accuracy')
plt.title('KNN Accuracy Comparison for Different Metrics (Bar Chart)')
plt.xticks(index + bar_width*1.5, k_values)
plt.ylim(0.8, 1.05)
plt.legend(title='Metric')
plt.grid(axis='y')
plt.show()
```
